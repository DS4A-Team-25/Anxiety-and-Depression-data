# filtering_viz function
**Makes sure that we can make different tables according to variables we want**

```python
df_viz = df.copy()

subgrp_variables = [subgroups for subgroups in df_viz['Subgroup']]
# state_variables = [state for state in df_viz['State']]

def filtering_viz(phase, indicator, group, subgroup = subgrp_variables):
    "This filters the data set by: Phase of Coronavirus Pandemic (1, 2, or 3), Indicator (Symptom of Anxiety Disorder or Symptoms of Depressive Disorder), Groups (By: Age, Education, Gender, Race/Hispanic ethnicity). Optional- Subgroup.  "
    filtered_table = df_viz[(df_viz['Phase'] == phase) & (df_viz['Indicator']== indicator) & (df_viz['Group'] == group) & (df_viz['Subgroup'] == subgroup)]
    return filtered_table
```

# Looking at Age

### Note the following:

**Percentages of people across age groups that DID NOT GET NEEDED CARE in the Last 4 Weeks in Phase 1**

```python
filtered_didnot_p1 = filtering_viz(1,'Did Not Get Needed Care, Last 4 Weeks', 'By Age')
# filtered_didnot_p1
# Examining Percentages of people across age groups that DID NOT GET NEEDED CARE in the Last 4 Weeks in Phase 1

viz_didnot_p1 = sns.catplot(x="Value", y="Subgroup", data= filtered_didnot_p1, kind = "violin", hue = 'State')
```
